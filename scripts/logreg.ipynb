```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python

dfs = [pd.read_csv(f'../data/freeAgents{year}.csv') for year in range(2011,2025)]
combined_df = pd.concat(dfs, ignore_index=True)

combined_df
```

```python
df_except_last = combined_df.iloc[:, 1:-1] 
scaler = StandardScaler()
df_normalized = df_except_last.copy()
df_normalized[df_except_last.columns] = scaler.fit_transform(df_except_last)

df_normalized
```

```python
y = combined_df['Cap']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df_normalized, y, test_size=0.2, random_state=42)


# Train the logistic regression model
# Initialize the classifier
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
scaler = StandardScaler()
df_career_normalized = df_career.copy()
df_career_normalized[df_career.columns] = scaler.fit_transform(df_career)
X_train_career, X_test_career, y_train_career, y_test_career = train_test_split(df_career_normalized, y, test_size=0.2, random_state=42)


# Train the logistic regression model
# Initialize the classifier
career_model = LinearRegression()

# Train the model
career_model.fit(X_train_career, y_train_career)

# Predict on the test set
y_pred_career = career_model.predict(X_test_career)

# Evaluate the model
mse = mean_squared_error(y_test_career, y_pred_career)
print(f'Mean Squared Error: {mse}')


career_sliced = career_predictions.iloc[:, 1:25]
career_data_normalized = scaler.fit_transform(career_sliced)
predictions = career_model.predict(career_data_normalized)
career_predictions['Prediction_Career'] = predictions
career_predictions['Prediction_Career'] = career_predictions['Prediction_Career'].apply(lambda x: max(x, 0))
career_predictions['Prediction_Career'] = career_predictions['Prediction_Career'].astype(int)
career_predictions
```
